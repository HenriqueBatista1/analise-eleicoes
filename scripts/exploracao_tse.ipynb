```python
# Votação 2018 - Candidatos por Município e Zona Eleitoral
caminho_arquivo = "../data/raw/votacao_candidato_munzona_2018_BR.csv"

# Lendo o CSV com os parâmetros corretos para o arquivo do TSE
df2 = pd.read_csv(caminho_arquivo, sep=';', encoding='latin-1')

# Exibindo as 5 primeiras linhas do dataframe
df2.head()
```

```python
# TABELA 1: Votos nominais por candidato no 1º turno
turno = 1

tabela = (
    df[df["NR_TURNO"] == turno]
    .groupby("NM_URNA_CANDIDATO", as_index=False)["QT_VOTOS_NOMINAIS"]
    .sum()
    .sort_values("QT_VOTOS_NOMINAIS", ascending=False)
)

# adiciona linha de soma
tabela.loc[len(tabela)] = {
    "NM_URNA_CANDIDATO": "SOMA",
    "QT_VOTOS_NOMINAIS": tabela["QT_VOTOS_NOMINAIS"].sum()
}

tabela
```

```python
# TABELA 2: Votos nominais por candidato no 2º turno
turno = 2

tabela = (
    df[df["NR_TURNO"] == turno]
    .groupby("NM_URNA_CANDIDATO", as_index=False)["QT_VOTOS_NOMINAIS"]
    .sum()
    .sort_values("QT_VOTOS_NOMINAIS", ascending=False)
)

# adiciona linha de soma
tabela.loc[len(tabela)] = {
    "NM_URNA_CANDIDATO": "SOMA",
    "QT_VOTOS_NOMINAIS": tabela["QT_VOTOS_NOMINAIS"].sum()
}

tabela
```

```python
# TABELA 3: Votos nominais por candidato no 1º turno, por estado
turno = 1

tabela = (
        df[df["NR_TURNO"] == turno]
        .pivot_table(
            index="SG_UF",
            columns="NM_URNA_CANDIDATO",
            values="QT_VOTOS_NOMINAIS",
            aggfunc="sum",
            fill_value=0
        )
        .sort_index()
)

# Soma de votos por estado
tabela["SOMA"] = tabela.sum(axis=1)

# Total nacional por candidato
tabela.loc["QT"] = tabela.sum()

tabela
```

```python
# TABELA 3: Votos nominais por candidato no 2º turno, por estado
turno = 2

tabela = (
        df[df["NR_TURNO"] == turno]
        .pivot_table(
            index="SG_UF",
            columns="NM_URNA_CANDIDATO",
            values="QT_VOTOS_NOMINAIS",
            aggfunc="sum",
            fill_value=0
        )
        .sort_index()
)

# Soma de votos por estado
tabela["SOMA"] = tabela.sum(axis=1)

# Total nacional por candidato
tabela.loc["QT"] = tabela.sum()

tabela
```

```python
# candidatos do segundo turno
candidatos_2t = (
    df.loc[df["NR_TURNO"] == 2, "NM_URNA_CANDIDATO"]
      .unique()
)

comparacao = (
    df[df["NM_URNA_CANDIDATO"].isin(candidatos_2t)]
    .groupby(["NM_URNA_CANDIDATO", "NR_TURNO"])["QT_VOTOS_NOMINAIS"]
    .sum()
    .unstack(fill_value=0)
)

comparacao.columns = ["QT_VOTOS_1T", "QT_VOTOS_2T"]

comparacao["DIFERENCA"] = (
    comparacao["QT_VOTOS_2T"] -
    comparacao["QT_VOTOS_1T"]
)

comparacao = (
    comparacao
    .reset_index()
    .sort_values("QT_VOTOS_2T", ascending=False)
)

comparacao
```
